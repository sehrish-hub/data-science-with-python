What is a List?
A list in python is a built-in collection type used to store multiple items in a single variable. You create it using square brackets [].

```python
#example
fruits = ["apple", "banana", "cherry"]
print(fruits)
```

                        Characteristics of lists
1. Ordered: items in a list are in a defined order and that order stays unless you modify the list.

```python
#example
#         0    1    2    3
numbers = [10, 20, 30, 40]
print(numbers)
print(numbers[0]) # Output: 10
print(numbers[2]) # Output: 30
```

2. Changeable / Mutable: You can change, add, or remove items after creating that list.

changing items: since list are mutable, you can assign a new value to a specific index.

```python
#example
colors = ["red", "green", "blue"]
colors[1] = "yellow"  # Changing the second item
print(colors)  # Output: ['red', 'yellow', 'blue']
# colors.remove("red")  # Removing an item
# print(colors)  # Output: ['yellow', 'blue']
```

Removing items:  you can remove items using remove(), pop(), del or clear().

```python
colors = ["red", "green", "blue"]
colors.remove("red")  # Removing an item
print(colors)  # Output: ['yellow', 'blue', 'purple']
```

```python
colors = ["red", "green", "blue"]
colors.remove("green")  # Removing an item
print(colors)  # Output: ['red', 'blue']
colors.pop(1)  # Removing the item at index 1
print(colors)  # Output: ['red']
```

```python
numbers = [10, 20, 30, 40, 50]
numbers.remove(30)  # Removing the item with value 30
print(numbers)  # Output: [10, 20, 40, 50]
numbers.pop(2)  # Removing the item at index 2 and returns it
print(numbers)  # Output: [10, 20, 50]
del numbers[0]  # Deleting the item at index 0
print(numbers)  # Output: [20, 50]
numbers.clear()  # Removing all items
print(numbers)  # Output: []
```

```python
colors = ["red", "green", "blue"]
colors.pop(1)  # Removing the item at index 1
print(colors)  # Output: ['red', 'blue']
```

Adding Items: You can add items to the end of the list using the append() method or insert them at a specific index using the insert() method, or extend() method to add multiple items etc.

```python
colors = ["red", "green", "blue"]
colors.append("purple")  # Adding a new item at the end
print(colors)  # Output: ['red', 'green', 'blue', 'purple']
colors.insert(1, "orange")  # Inserting an item at index 1
print(colors)  # Output: ['red', 'orange', 'green', 'blue', 'purple']
```

```python
colors = ["red", "green", "blue"]
colors.insert(1, "orange")  # Inserting an item at index 1
print(colors)  # Output: ['red', 'orange', 'green', 'blue']
```

3. Allow Duplicates: List can have duplicate values.

```python
#example
items = ["apple", "mango", "cherry", "orange", "cherry"]
print(items)  # Output: ['apple', 'mango', 'cherry', 'orange', 'cherry']
```

4. Mixed Data types: Items in a list can be different data types(strings, integers, booleans etc.)

```python
mixed_list = ["hello", 42, 3.14, True]
print(mixed_list)  # Output: ['hello', 42, 3.14, True]
```

How to Create a List?
you create a list by putting values in square brackets [], separated by commas. you can also use the list() constructor.


```python
#example (literal method)
foods = ["pizza", "burger", "pasta"]
print(foods)  # Output: ['pizza', 'burger', 'pasta']
```

```python
#example (list() constructor method)
foods = list(("pizza", "burger", "pasta")) #you can create a list using the list() constructor by passing a tuple.
print(foods)  # Output: ['pizza', 'burger', 'pasta']
```

Accessing List Elements (list k items tak pohnchna)?
Each item in a list has an index: the first item is index 0, second is 1 , and so on. you can use negative indexing(-1, -2,..) to access items from the end.

```python
#example 
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[0])  # Output: Python
print(languages[3:5])  # Output: ['JavaScript', 'Ruby']
```

```python
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[:3]) # Output: ['Python', 'Java', 'C++']
```

```python
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[4:]) # Output: ['Ruby', 'Go', 'Swift']
```

```python
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[-1])  # Output: Swift
```

```python
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[-3:-1])  # Output: ['Ruby', 'Go']
```

```python
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[:-2])  # Output: ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
print(languages[-5:])  # Output: ['C++', 'JavaScript',
```

```python
languages = ["Python", "Java", "C++", "JavaScript", "Ruby", "Go", "Swift"]
print(languages[-3:6])  # Output: ['Ruby', 'Go']
```

Slicing(hissy): you can get a subset of the list by slicing: list[start:stop]. The stop index is excluded.

```python
#example
mylist= ["a", "b", "c", "d", "e", "f", "g"]
print(mylist[1:5])  # Output: ['b', 'c', 'd', 'e'] list k index 1 sy 4 tak (5 excluded)
print(mylist[:4])   # Output: ['a', 'b', 'c', 'd'] start sy 3 tak (4 excluded)
print(mylist[3:])   # Output: ['d', 'e', 'f ', 'g'] index 3 sy end tak
```

List length & Type:
use len(list) to get number of items.
use type(list) to check it's a list object.

```python
flower = ["rose", "tulip", "daisy", "sunflower", "lily"]
print(len(flower))  # Output: 5
print(type(flower)) # Output: <class 'list'>
```
