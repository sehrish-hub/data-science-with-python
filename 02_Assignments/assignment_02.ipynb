## 7) Compound Interest
Write a Python program to calculate compound interest. 
# Formula:
#              A = P * (1 + r / n) ^ (n * t)
Where: 
A= total amount 
P = principal amount 
r = annual interest rate (decimal) 
n = number of times interest is compounded per year 
t = time in years 
Take P, r, n, and t as inputs from the user. 

```python
P = float(input("Enter principal amount : "))
r = float(input("Enter annual interest rate (in %): "))
# Convert percentage to decimal
r = r / 100
t = float(input("Enter time (in years): "))
n = float(input("Enter number of times interest is compounded per year: "))
A = P * (1 + r / n) ** (n * t) # Amount after interest
print(f"Amount after {t} years: {A:.2f}")
CI = A - P # Compound Interest
print(f"Compound Interest = {CI:.2f}")
```

## 11) Quadratic Equation Roots

```python
# Quadratic Equation Roots
print("For the quadratic equation ax^2 + bx + c = 0, enter the coefficients:")
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
# Calculate discriminant
D = b**2 - 4*a*c
# Calculate roots
root1= x1=(-b + D**0.5) / (2*a)
root2= x2=(-b - D**0.5) / (2*a)
print(f"Roots of the quadratic equation are:\n x1 = {root1:.2f}\n x2 = {root2:.2f}")
```

## 13) Gravitational Force
Write a Python program to calculate the gravitational force between two objects. 
# Formula: 
#             F = G(m1m2/r^2)         
m1, m2 are the masses of the objects 
r = distance between the centers of the objects 
Take m1, m2 and r as inputs from the user. 

```python
m1 = float(input("Enter mass of first object (in kg): "))
m2 = float(input("Enter mass of object (in kg): "))
r = float(input("Enter distance between the centers of the objects (in meters): "))
G = 6.67e-11  # Gravitational constant
F = G * (m1 * m2) / (r ** 2)
print(f"Gravitational Force = {F:.6e} N")
```
