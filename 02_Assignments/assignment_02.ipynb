## 1) Area of Rectangle
Write a Python program to calculate the area of a rectangle.
# Formula:
#            Area = length Ã— width
Take length and width as inputs from the user.

```python
length = float(input("Enter length: "))
width = float(input("Enter width: "))
Area = length * width
print(f"Area = {Area:.2f}")
```

## 2) Circumference of Circle
Write a Python program to calculate the circumference of a circle.
# Formula:    
#             Circumference = 2ğ…r 
Take the radius r as input from the user.

```python
r = float(input("Enter radius: "))
circumference = 2 * 3.14 * r
print(f"Circumference = {circumference:.3f}")
```
