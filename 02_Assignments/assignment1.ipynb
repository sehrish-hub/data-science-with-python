```python
# Question 1 — Swap Two Variables
# Ask the user for 2 numbers and swap them without using a third variable.
a = int(input("Enter A: "))
b = int(input("Enter B: "))
print("Before swapping: A =", a, "B =", b) 
# swapping without third variable
a = a + b 
b = a - b 
a = a - b 
print("After swapping: A =", a, "B =", b)
```

```python
# Question 2 — Find the Last Digit
# Ask the user for a number and print its last digit.
num = int(input("Enter a number: ")) #
last_digit = num % 10 # Getting the last digit using modulus operator
print("The last digit is:" , last_digit)
```

```python
# Question 3 — Length of String × 2
# Ask the user for a string. Print the length of the string multiplied by 2.
user_string = input("Enter a string: ") # Getting user input
length_times_two = len(user_string) * 2 # Calculating length multiplied by 2
print("Length of the string multiplied by 2 is:", length_times_two)
```

```python
# Question 4 — Name + Age Calculation
# Ask the user for their name and age. Print a sentence adding 5 to their age.
name = input("Enter your name: ") # Getting user's name
age = int(input("Enter your age: ")) # Getting user's age
new_age = age + 5 # Adding 5 to the age
print("Hello", name + ", in 5 years you will be", new_age, "years old.")
```

```python
# Question 5 — Sum of Digits (3-digit number)
# Ask the user for a 3-digit number and print the sum of its digits.
three_digit_num = int(input("Enter a 3-digit number: ")) # Getting user input
hundreds = three_digit_num // 100 # Extracting the hundreds digit (Example: 753 // 100 = 7)
tens = (three_digit_num // 10) % 10 # Extracting the tens digit (753 // 10 = 75 → 75 % 10 = 5)
units = three_digit_num % 10 # Extracting the units digit (753 % 10 = 3)
sum_of_digits = hundreds + tens + units # Calculating the sum of digits
print("The sum of the digits is:", sum_of_digits)
```

```python
# Question 6 — Add First & Last Digit
# Ask the user for a 3-digit number and print the sum of the first and last digits.
three_digit_num = int(input("Enter a 3-digit number: ")) # Getting user input
first_digit = three_digit_num // 100 # Extracting the first digit (Example: 753 // 100 = 7)
last_digit = three_digit_num % 10 # Extracting the last digit (753 % 10 = 3)
sum_first_last = first_digit + last_digit # Calculating the sum of first and last digits
print("The sum of the first and last digits is:", sum_first_last)
```

```python
# Question 7 — Repeat User’s Name
# Ask the user for their name. print their name 5 times using string multiplication.
name = input("Enter your name: ") # Getting user's name
repeated_name = ((name + " ") * 5) # Repeating the name 5 times using string multiplication
print("Your name repeated 5 times is:", repeated_name)
```

```python
# Question 8 — Convert Minutes to Seconds
# Ask the user for minutes and convert it to seconds.
minutes = int(input("Enter minutes: ")) # Getting user input in minutes
seconds = minutes * 60 # Converting minutes to seconds
print(minutes, "minutes is equal to", seconds, "seconds.")
```

```python
# Question 9 — Average of Three Numbers
# Ask the user for 3 numbers and print their average.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))
average = (num1 + num2 + num3) / 3
print("The average of the three numbers is:", average)
```

```python
# Question 9 — Extract second digit
# Ask the user for a 3-digit number and print the middle digit using only operators.
three_digit_num = int(input("Enter a 3-digit number: ")) # Getting user input
middle_digit = (three_digit_num // 10) % 10 # Extracting the middle
print("The middle digit is:", middle_digit)
```
