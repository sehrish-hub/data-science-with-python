Student Report Card System

```python
# -------------------------------
#       STUDENT REPORT CARD
# -------------------------------
print("=== STUDENT REPORT CARD SYSTEM ===")

#  1. Taking Inputs from User
name = input("Enter Student Name: ")
father_name = input("Enter Father's Name: ")
student_class = input("Enter Class: ")

#  2. Taking Marks for 5 Subjects
eng = int(input("Enter English Marks: "))
math = int(input("Enter Math Marks: "))
science = int(input("Enter Science Marks: "))
computer = int(input("Enter Computer Marks: "))
urdu = int(input("Enter Urdu Marks: "))

#  3. Calculating Total Marks and Percentage
total_marks = 500
obtained_marks = eng + math + science + computer +urdu
percentage = (obtained_marks / total_marks) * 100

# 4. Roll Number generation
# # First letter of Name + Last letter of Father Name + Class
roll_number = name[0].upper() + father_name[-1].upper() + str(student_class)

#  5. Displaying Report Card
print("\n----- REPORT CARD -----"),
print(f"Name           : {name}")
print(f"Father's Name  : {father_name}")
print(f"Class          : {student_class}")
print(f"Roll Number    : {roll_number}")
print("----------------------------------")
print(f"English        : {eng}")
print(f"Math           : {math}")
print(f"Science        : {science}")
print(f"Computer       : {computer}")
print(f"Urdu           : {urdu}")
print("----------------------------------")
print(f"Total Marks    : {obtained_marks} / {total_marks}")
print(f"Percentage     : {percentage:.2f}%")
print("----------------------------------")
```

```python
import math

# Take inputs from the user
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Compute the discriminant
discriminant = b**2 - 4*a*c

# Check if roots are real or complex
if discriminant >= 0:
    # Real roots
    root1 = (-b + math.sqrt(discriminant)) / (2*a)
    root2 = (-b - math.sqrt(discriminant)) / (2*a)
    print(f"The roots are real: {root1} and {root2}")
else:
    # Complex roots
    real_part = -b / (2*a)
    imaginary_part = math.sqrt(abs(discriminant)) / (2*a)
    print(f"The roots are complex: {real_part} + {imaginary_part}i and {real_part} - {imaginary_part}i")
```
